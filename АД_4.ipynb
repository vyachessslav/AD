```python
import numpy as np
import random
from math import comb

num_trials = 1000000
```

Задача 2.7. Бросают 4 игральные кости. Найти вероятность того, что на них выпадет одинаковое число очков.

```python
dice_rolls = np.random.randint(1, 7, size=(num_trials, 4))

all_same = np.all(dice_rolls == dice_rolls[:, [0]], axis=1)
prob_monte_2_7 = np.mean(all_same)

prob_exact_2_7 = 6 / (6**4)
print(f"Монте-Карло (numpy): {prob_monte_2_7:.6f}")
print(f"Точное значение: {prob_exact_2_7:.6f}")
```



Задача 2.8. В коробке 10 красных и 6 синих пуговиц. Какова вероятность того, что две наудачу вынутые пуговицы будут одноцветными?

```python
buttons_array = np.array([0]*10 + [1]*6)

favorable = 0
for _ in range(num_trials):
    # Перемешиваем пуговицы
    np.random.shuffle(buttons_array)
    # Берем первые две пуговицы
    if buttons_array[0] == buttons_array[1]:
        favorable += 1

prob_monte_2_8 = favorable / num_trials
prob_exact_2_8 = (comb(10, 2) + comb(6, 2)) / comb(16, 2)

print(f"Монте-Карло: {prob_monte_2_8:.6f}")
print(f"Точное значение: {prob_exact_2_8:.6f}")
```

Задача 3.8. Стержень длины l разломан в двух наудачу выбранных точках. С какой вероятностью из полученных отрезков можно составить треугольник?

```python
x = np.random.rand(num_trials, 2)
# Сортируем, чтобы первая точка была левее второй
x.sort(axis=1)
a = x[:, 0]
b = x[:, 1] - a
c = 1 - x[:, 1]

cond = (a + b > c) & (a + c > b) & (b + c > a)
prob = np.mean(cond)

print(f"Монте-Карло: {prob:.6f}")
print(f"Точное значение: 0.25")
```
