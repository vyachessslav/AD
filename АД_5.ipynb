```python
import pandas as pd


data = pd.read_csv("beverage_r.csv", sep=";")
```

Задача 1

```python
from sklearn.model_selection import train_test_split


train_data, test_data = train_test_split(data, test_size=0.33, random_state=42)

print('Длина тренировочного набора:', len(train_data))
print('Длина тестового набора:', len(test_data))
```

Задача 2

```python
data = data.sample(frac=1, random_state=42).reset_index(drop=True)

part_size = len(data) // 4
parts = [data.iloc[i*part_size:(i+1)*part_size] for i in range(4)]
for i in range(len(data) % 4):
    parts[i] = pd.concat([parts[i], data.iloc[part_size*4 + i:part_size*4 + i + 1]])

for part in parts:
    print(len(part))
```
